```python
import pandas as pd
```

```python
#read csv file already gathered through web scraping
df = pd.read_csv("./stats.csv")
```

```python
#start dropping irrelevant columns
df = df.drop("G+A", axis=1)
```

```python
df = df.drop("G-PK", axis = 1)
```

```python
df = df.drop("PKatt", axis = 1)
```

```python
#drop columns irrelevant to the context of our project
df.drop(["npxG+xAG", "PrgR", "Gls.1", "Ast.1","G+A.1","G-PK.1","G+A-PK","xG.1","xAG.1","xG+xAG","npxG.1","npxG+xAG.1", "Matches"], axis = 1, inplace=True)
```

```python
columns=['Team','Age', 'MP', 'Starts', 'Min', '90s', 'Gls', 'Ast', 'PK', 'CrdY', 'CrdR', 'xG', 'npxG', 'xAG', 'PrgC', 'PrgP']
teams = pd.DataFrame(columns=columns)

#separate data from teams into another data frame so it can be represented as an entity
for index, row in df.iterrows():
    if row['Player'] == 'Squad Total':
        team = [row['Team'], row['Age'], row['MP'], row['Starts'], row['Min'], row['90s'], row['Gls'], row['Ast'], row['PK'], row['CrdY'], row['CrdR'], row['xG'],row['npxG'],row['xAG'], row['PrgC'],row['PrgP']]
        new_row = pd.DataFrame([team], columns=columns)
        teams = pd.concat([teams if not teams.empty else None, new_row], ignore_index=True)
    
```

```python
teams
```

```python
teams = teams.set_index('Team')
```

```python
teams.to_csv('./teams.csv')
```
